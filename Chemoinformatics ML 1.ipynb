```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
```

```python
activity = new_client.activity
act_res = activity.filter(target_chembl_id="CHEMBL4523582", standard_type="IC50") 
# Target CHEMBL ID - CHEMBL4523582 # 3C like protease or Mpro
# assay_organism="Homo sapiens"
# assay_chembl_id="CHEMBL4495582" # FRET Assay 
# assay_chembl_id="CHEMBL4495583" # FRET Assay 
```

```python
act_df = pd.DataFrame(act_res)
act_df
```

```python
# Save csv cache
act_df.to_csv("./output/sars_cov2_mpro_ic50.csv")
```

# Clean and Prepare a subset of dataset

```python
act_df.head(3)
```

```python
act_df.columns
```

```python
act_df[["value", "upper_value", "units", "standard_type", "standard_value", "standard_units", "standard_relation"]].head()
```

### Remove non equal relations and and duplicate molecules

```python
sub_df = act_df[act_df["standard_relation"] == "="]
sub_df.shape
```

```python
sub_df = sub_df.sort_values("molecule_chembl_id").drop_duplicates(["molecule_chembl_id"], keep="last")
```

```python

print("Molecule Count", sub_df.molecule_chembl_id.nunique())
# Unique Relations
print("Relation Count", sub_df.relation.nunique())
print("Standard Relation Count", sub_df.standard_relation.nunique())
print("Unit Count", sub_df.units.nunique())
print("Standard Unit Count", sub_df.standard_units.nunique())
```

```python
sub_df.shape
```

### Create subset

```python
sub_df = sub_df[["molecule_pref_name", "canonical_smiles", "molecule_chembl_id", "standard_value", "standard_units"]]
sub_df.head()
```

## Convert data types and create a ID column

```python
# Get Integer Molecule ID
sub_df['molecule_id'] = sub_df['molecule_chembl_id'].map(lambda x: x.replace("CHEMBL", ""))
sub_df.head()
```

```python
numeric_columns = ["molecule_id", "standard_value"]
sub_df[numeric_columns] = sub_df[numeric_columns].apply(pd.to_numeric, errors='coerce')
sub_df.dtypes
```

```python
sub_df.molecule_id.nunique()
```

### Quick Null Check

```python
sub_df['canonical_smiles'].isnull().sum()
```

```python
sub_df['standard_value'].isnull().sum()
```

```python
# Remove Null columns
sub_df = sub_df.dropna(subset=['canonical_smiles', 'standard_value'])
print(sub_df['canonical_smiles'].isnull().sum())
sub_df.shape
```

# Write Clean dataset to CSV

```python
sub_df.reset_index(drop=False).to_csv("./output/sars_cov2_mpro_ic50_clean.csv")
```
