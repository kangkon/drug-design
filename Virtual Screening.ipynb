```python
import os
import pandas as pd

RECEPTOR = "7BQY"

# Receptor filename naming convention - RECEPTOR_filename.pdbqt

WORK_DIR = "/mnt/d/computers-in-biology/docking"
LIG_DIR = os.path.join(WORK_DIR, "ligands_pdbqt")
RECEPTOR_DIR = os.path.join(WORK_DIR, "receptors", RECEPTOR)
OUT_DIR = os.path.join(WORK_DIR, "runs2", RECEPTOR)
```

### 1. List PDBQT files from ligand dir

```python
# List ligands
ligands = []
for file in os.listdir(LIG_DIR):
    if(os.path.isfile(os.path.join(LIG_DIR, file)) & file.endswith('.pdbqt')):
        ligands.append(file)
```

### 2. Create and process output folders

```python
from shutil import copy2
```

```python
prepared_lig_dirs = []
try:
    #if(os.path.isdir(OUT_DIR)):
    #    os.rmtree(OUT_DIR)
    
    for lig in ligands:
        lig_filename = os.path.splitext(lig)[0] # Ligfilename without extension
        out_dir_lig = os.path.join(OUT_DIR, lig_filename) # Output directory for current ligand
        
        # Ceate output dir for current ligand
        os.makedirs(out_dir_lig, exist_ok=True)
        
        # Copy Receptor and conf to Output directory for current ligand
        copy2(os.path.join(RECEPTOR_DIR, f"{RECEPTOR}_final.pdbqt"), out_dir_lig) # Receptor
        copy2(os.path.join(RECEPTOR_DIR, "conf.txt"), out_dir_lig) # Conf
        
        # Copy the PDBQT ligand file to destination
        copy2(os.path.join(LIG_DIR, lig), out_dir_lig)
        
        # Keep track of prepared ligands directories
        prepared_lig_dirs.append(out_dir_lig)
except OSError as error:
    print(error)
    print("Directory '% s' can not be removed" % OUT_DIR)
```

## Run Vina Docking to generate results and log

```python
import shlex, subprocess 
from datetime import datetime
```

```python
def run_docking(lig_out_dir, output_logs):
    """
    Run Docking
    lig_out_dir: Output directory for a ligand
    """
    
    print(f"Running {lig_out_dir}")
    
    # Write Log
    output_logs = output_logs + f"\n[STARTRUN] {datetime.now()} OUTDIR {lig_out_dir}\n[STARTLOG]\n"
    
    ligand = f"{os.path.basename(lig_out_dir)}.pdbqt"
    #args = shlex.split(f"vina --help")
    args = shlex.split(f"vina --config conf.txt  --ligand {ligand}")
    process = subprocess.Popen(
        args, 
        stdout=subprocess.PIPE, 
        stderr=subprocess.PIPE,
        cwd=os.path.join(lig_out_dir) # Set current working directory
    )
    output, error = process.communicate()
    if error:
        print("Error: ", error) 
        output_logs = output_logs + error.decode("utf-8")
    else:
        print(output.decode("utf-8"))
        output_logs = output_logs + output.decode("utf-8") 
    output_logs = output_logs + f"\n[ENDLOG]\n[ENDRUN] {datetime.now()}\n++++++++++\n"
    
    print("\n++++++++++\n")
    return output_logs

output_logs = ""
for i in prepared_lig_dirs:
    output_logs = run_docking(i, output_logs)
```

```python
f = open("./output/docking_output.txt", "a")
f.write(output_logs)
f.close()
```
